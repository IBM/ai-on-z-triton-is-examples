## Load data

```python
train_loader, test_loader, train_labels, test_labels = load_data()
```

```python
train_steps = len(train_loader)
test_steps = len(test_loader)
```

## Build and Train Model

```python
model_inceptionV3 = build_model_InceptionV3()
model_inceptionV3.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics = ['accuracy'])
model_inceptionV3.summary()
```

```python
history_train =  model_inceptionV3.fit( train_loader,
                                        steps_per_epoch=train_steps,
                                        epochs =10,
                                        validation_data=test_loader,
                                        validation_steps=test_steps
                                      )
```

```python
model_inceptionV3.save("models/inception-v3/inception-v3.h5")

with open("modelsinception-v3/inception-v3_history.pkl", 'wb') as fp:
    pickle.dump(history_train.history, fp)
```

```python
print("Inception v3 :\n", model_inceptionV3.evaluate(test_loader))
```
